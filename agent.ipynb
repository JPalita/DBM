```python
import numpy as np
```

```python
class Agent:
    def __init__(self, a0, b0, q):
        self.a0 = a0
        self.b0 = b0
        self.q = q
        self.p = 0.0
        self.pickP()
        self.notPickP()
        
    def pickP(self):#PICK P  FROM BETA DISTRIBUTION
        self.p = np.random.beta(self.a0, self.b0)
    def notPickP(self):
        self.p = 1 - np.random.beta(self.a0, self.b0)
            
    def draw(self):
        if np.random.uniform(0,1) < self.p:
            return True
        else:
            return False
        
    def turn(self):
        if np.random.uniform(0,1) < self.q:
            self.pickP()
        # else:
        #     self.notPickP()
        return self.draw()
```

```python
ag = Agent(0.1,0.5, 0.1)
n = 8
outcomes = []
prob = []
for i in range(n):
    outcomes.append(ag.turn())
```
